```python
# Here I am using the 2018-2019 NBA Player Stats to cluster players according to their miutes played and points per game
# MPG = Minutes Per Game
# PPG = Points Per Game
```

```python
import numpy as np
from pandas import DataFrame
from pandas import read_excel
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder

# imported data from excel file
my_sheet = "Sheet1"
file_name = "2018-2019NBAPlayerStats.xlsx"
Data = read_excel(file_name, sheet_name = my_sheet)
# print(Data.head()) # testing import

df = Data[["MPG", "PPG"]]

kmeans = KMeans(n_clusters = 5).fit(df)
centroids = kmeans.cluster_centers_
print(centroids)

plt.scatter(Data["MPG"], Data["PPG"], c=kmeans.labels_, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red')

plt.show()
```
