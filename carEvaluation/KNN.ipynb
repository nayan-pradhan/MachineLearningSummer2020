```python
# using KNN

import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing

data = pd.read_csv("car.data")
print(data.head())

# we have to convert the text strings (non numerical data) into integer values that corresponds to our string
# we use import sklearn to do this

le = preprocessing.LabelEncoder() # takes the labels and encodes them into appropriate integer values
buying = le.fit_transform(list(data["buying"]))
maint = le.fit_transform(list(data["maint"]))
door = le.fit_transform(list(data["door"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))

print(buying) # example of how encoding works

predict = "class"

X = list(zip(buying, maint, door, persons, lug_boot, safety)) # converts all these into one big tuple list
y = list(cls)

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.1)

model = KNeighborsClassifier(n_neighbors = 6)

model.fit(x_train, y_train)
acc = model.score(x_test, y_test)
print(acc) # prints the accuracy

predicted = model.predict(x_test)
names = ["unacc", "acc", "good", "vgood"]

for x in range(len(predicted)):
    print("Predicted: ", names[predicted[x]], "| Data: ", x_test[x], "| Actual: ", names[y_test[x]])
```
