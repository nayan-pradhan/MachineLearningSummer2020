```python
# Import
import os
import nltk
import nltk.corpus
```

```python
# Specify Website 
import urllib.request
response = urllib.request.urlopen('https://www.apple.com/iphone-11-pro/')
html = response.read()
print(html)
```

```python
# Remove all html5 elements form website using BeautifulSoup
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html5lib')
text = soup.get_text(strip = True)
print(text)
```

```python
# Convert the words into individual tokens
tokens = [t for t in text.split()]
print(tokens)
```

```python
# Removes stop words like 'a', 'at', 'the', etc
from nltk.corpus import stopwords
sr = stopwords.words('english')
clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('english'):
        clean_tokens.remove(token)
        
# Frequency distribution for tokens
freq = nltk.FreqDist(clean_tokens)
for key, val in freq.items():
    print(str(key) + ":" + str(val))

# Plot
freq.plot(20, cumulative = False)
```
